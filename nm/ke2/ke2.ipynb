# Fixed-point iteration for $x = \exp(x)$

In this notebook we investigate the equation

$$
x = e^x.
$$

A fixed-point problem has the form

$$
x = g(x),
$$

and the simplest fixed-point iteration is

$$
x_{n+1} = g(x_n).
$$

For this problem the natural choice is

$$
g(x) = e^x,
$$

so the iteration becomes

$$
x_{n+1} = e^{x_n}.
$$


## First check: does a real solution exist?

Before iterating, define

$$
f(x) = e^x - x.
$$

A real solution of $x=e^x$ would be a zero of $f$. But

$$
f'(x) = e^x - 1.
$$

The only critical point is at $x=0$, and

$$
f(0) = 1.
$$

Since this critical point is the global minimum, $f(x)>0$ for every real $x$. Therefore:

**The equation $x=e^x$ has no real solution.**

The iteration below is still useful because it shows what goes wrong numerically.

```python
from math import exp, isfinite


def fixed_point_exp(x0, max_iter=20):
    """Run x_{n+1} = exp(x_n) starting from x0."""
    values = [float(x0)]

    for _ in range(max_iter):
        try:
            x_next = exp(values[-1])
        except OverflowError:
            values.append(float("inf"))
            break

        values.append(x_next)

        if not isfinite(x_next):
            break

    return values
```

## Try the iteration

Let us start with a few different initial values. If the method converged to a real fixed point, the sequence would settle down to a finite number.

```python
starts = [-2, -1, 0, 0.5, 1]

for x0 in starts:
    values = fixed_point_exp(x0, max_iter=10)
    print(f"x0 = {x0}")
    for n, x in enumerate(values):
        print(f"  n={n:2d}: x_n = {x:.12g}")
    print()
```

The sequence does not converge. Even when we start with a negative value, one application of $e^x$ makes the next iterate positive. After that, the values grow very quickly.

This is not an accident. Since $e^x>x$ for every real $x$, each fixed-point step produces a value above the previous one once the sequence is positive, and there is no finite real fixed point for it to approach.

## A convergence check

A common local convergence condition for fixed-point iteration is

$$
|g'(x^*)| < 1,
$$

where $x^*$ is the fixed point. Here

$$
g'(x) = e^x.
$$

For real $x$, $e^x$ is always positive. But the more fundamental issue is that there is no real $x^*$ satisfying $x^*=e^{x^*}$ in the first place.

## Numerical conclusion

The fixed-point iteration for

$$
x = e^x
$$

does **not** calculate a real solution, because no real solution exists. Numerically, the natural iteration

$$
x_{n+1} = e^{x_n}
$$

diverges rapidly.

So the correct real-valued result is:

$$
\boxed{\text{No real solution exists.}}
$$


## Optional comparison: a similar equation that does converge

A nearby equation that is often used in fixed-point iteration examples is

$$
x = e^{-x}.
$$

That equation has a real solution, and the fixed-point iteration

$$
x_{n+1}=e^{-x_n}
$$

does converge from many starting values.

```python
def fixed_point_exp_minus(x0, tolerance=1e-12, max_iter=100):
    """Run x_{n+1} = exp(-x_n) until consecutive iterates are close."""
    x = float(x0)

    for n in range(1, max_iter + 1):
        x_next = exp(-x)

        if abs(x_next - x) < tolerance:
            return x_next, n

        x = x_next

    return x, max_iter


root, iterations = fixed_point_exp_minus(x0=0)

print(f"Approximate solution of x = exp(-x): {root:.12f}")
print(f"Iterations needed: {iterations}")
print(f"Residual exp(-x) - x: {exp(-root) - root:.3e}")
```
